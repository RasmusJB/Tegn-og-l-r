```python
# create a list of random fruits
fruits = [
    "Apple",
    "Banana",
    "Orange",
    "Strawberry",
    "Blueberry",
    "Mango",
    "Pineapple",
    "Watermelon",
    "Grapes",
    "Kiwi"
]

print("List of random fruits:")
for fruit in fruits:
    print(f"- {fruit}")
```

```python
# create a dictionary with fruits and their colors
fruits_with_colors = {
    "Apple": "Red",
    "Banana": "Yellow",
    "Orange": "Orange",
    "Strawberry": "Red",
    "Blueberry": "Blue",
    "Mango": "Yellow",
    "Pineapple": "Yellow",
    "Watermelon": "Green",
    "Grapes": "Purple",
    "Kiwi": "Green"
}

print("Fruits with their colors:")
for fruit, color in fruits_with_colors.items():
    print(f"- {fruit}: {color}")
```

```python
# choose a random color and remove all fruits with that color
random_color = random.choice(list(fruits_with_colors.values()))

# find all fruits with the chosen color
removed_fruits = [fruit for fruit, color in fruits_with_colors.items() if color == random_color]

# remove them from the dictionary
for fruit in removed_fruits:
    del fruits_with_colors[fruit]

print(f"Eliminated color: {random_color}")
print(f"Removed fruits: {removed_fruits}")
print(f"\nRemaining fruits with colors:")
for fruit, color in fruits_with_colors.items():
    print(f"- {fruit}: {color}")
```

```python
# show what fruits and colours were removed
print("=" * 50)
print("ITEMS ELIMINATED")
print("=" * 50)
print(f"Color eliminated: {random_color}")
print(f"\nFruits removed ({len(removed_fruits)} total):")
for fruit in removed_fruits:
    print(f"  - {fruit} ({random_color})")
print("=" * 50)
```
