```python
class Person:
    def __init__(self, name, age, home_location=None):
        self.name = name
        self.age = age
        # home_location should be a tuple: (latitude, longitude)
        if home_location is None:
            # Default location (0.0, 0.0)
            self.home_location = (0.0, 0.0)
        else:
            self.home_location = home_location

    def happyBirthday(self):
        self.age += 1

person1 = Person("Alice", 30, home_location=(55.6761, 12.5683))  # Copenhagen
person2 = Person("Bob", 25, home_location=(48.8566, 2.3522))     # Paris
print(person1.name, person1.age, person1.home_location)
print(person2.name, person2.age, person2.home_location)
person1.happyBirthday()
print(person1.name, person1.age, person1.home_location)
```

```python
# create a list of ten random names
random_names = [
    "Alice",
    "Bob",
    "Charlie",
    "Diana",
    "Ethan",
    "Fiona",
    "George",
    "Hannah",
    "Isaac",
    "Julia"
]

print("List of ten random names:")
for name in random_names:
    print(f"- {name}")
```

```python
import random

# Center of Roskilde, Denmark
roskilde_lat = 55.6415
roskilde_lon = 12.0803

persons = []
for _ in range(10):
    name = random.choice(random_names)
    age = random.randint(19, 29)
    # Randomize location within ~0.01 degrees of center
    lat = roskilde_lat + random.uniform(-0.01, 0.01)
    lon = roskilde_lon + random.uniform(-0.01, 0.01)
    person = Person(name, age, home_location=(lat, lon))
    persons.append(person)

for p in persons:
    print(f"{p.name}, age {p.age}, location: {p.home_location}")
```

```python
import random

# Center of Roskilde, Denmark
roskilde_lat = 55.6415
roskilde_lon = 12.0803

# Copy the random_names list to avoid modifying the original
available_names = random_names.copy()
persons = []

def get_title(age):
    if age < 24:
        return "young"
    elif age > 25:
        return "old"
    else:
        return ""

for _ in range(10):
    if not available_names:
        # If we run out of names, generate a new random name
        name = f"Person{random.randint(1000, 9999)}"
    else:
        name = random.choice(available_names)
        available_names.remove(name)
    age = random.randint(19, 29)
    title = get_title(age)
    lat = roskilde_lat + random.uniform(-0.01, 0.01)
    lon = roskilde_lon + random.uniform(-0.01, 0.01)
    person = Person(f"{name} ({title})" if title else name, age, home_location=(lat, lon))
    persons.append(person)

for p in persons:
    print(f"{p.name}, age {p.age}, location: {p.home_location}")
```

```python
import matplotlib.pyplot as plt

# Extract latitudes and longitudes from persons
lats = [p.home_location[0] for p in persons]
lons = [p.home_location[1] for p in persons]
names = [p.name for p in persons]

# Center of Roskilde
center_lat = 55.6415
center_lon = 12.0803

plt.figure(figsize=(8, 8))
plt.scatter(lons, lats, c='blue', label='Person Home')
plt.scatter(center_lon, center_lat, c='red', marker='*', s=200, label='Roskilde Center')

# Annotate each person
for name, lat, lon in zip(names, lats, lons):
    plt.text(lon + 0.001, lat, name, fontsize=9)

plt.title('Home Locations of People Around Roskilde')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```
